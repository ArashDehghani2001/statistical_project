```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 13

n = 150
p = 0.2
graphs = []
for i in range(100):
    graphs.append(nx.fast_gnp_random_graph(n, p))
isolates = 0
connected = 0
for g in graphs:
    isolates += len(list(nx.isolates(g))) / n
    connected += int(nx.is_connected(g))
print("probability of being isolated => {}".format(isolates/100))
print("probability of being connected => {}".format(connected/100))
```

```python
# 14

ns = [int(i) for i in np.linspace(10, 150, 15)]
not_isolated = []
connected = []
for n in ns:
    p = 4 * np.log(n) / n
    graphs = []
    for i in range(100):
        graphs.append(nx.fast_gnp_random_graph(n, p))
    not_isolates = 0
    connects = 0
    for g in graphs:
        not_isolates += 1 - len(list(nx.isolates(g)))/n
        connects += int(nx.algorithms.components.is_connected(g))
    not_isolated.append(not_isolates/100)
    connected.append(connects/100)
plt.figure()
plt.title("probability of not being isolated")
plt.plot(ns,not_isolated)
plt.show()
plt.figure()
plt.title("probability of being connected(for graph)")
plt.plot(ns,connected)
plt.show()
```

```python
# 15

ns = [int(i) for i in np.linspace(10, 150, 15)]
not_isolated = []
connected = []
for n in ns:
    p = 4 / n
    graphs = []
    for i in range(100):
        graphs.append(nx.fast_gnp_random_graph(n, p))
    not_isolates = 0
    connects = 0
    for g in graphs:
        not_isolates += 1 -  len(list(nx.isolates(g))) / n
        connects += int(nx.algorithms.components.is_connected(g))
    not_isolated.append(not_isolates/100)
    connected.append(connects/100)
plt.figure()
plt.plot(ns,not_isolated)
plt.title("probability of not being isolated")
plt.show()
plt.figure()
plt.plot(ns,connected)
plt.title("probability of being connected(for graph)")
plt.show()
```
