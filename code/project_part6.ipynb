```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 9

n = 100
p= 0.34
graphs = []
for i in range(100):
    graphs.append(nx.fast_gnp_random_graph(n, p))
triangles = []
for g in graphs:
    triangles.append(sum(nx.triangles(g).values())/3)
print("triangles => {}".format(sum(triangles)/100))
```

```python
#10

ns = [int(i) for i in np.linspace(10, 100, 10)]
y = []
for n in ns:
    p = 60 / (n ** 2)
    graphs = []
    for i in range(100):
        graphs.append(nx.fast_gnp_random_graph(n, p))
    triangles = []
    for g in graphs:
        triangles.append(sum(nx.triangles(g).values()) / 3)
    y.append(sum(triangles) / 100)
print(y)
plt.figure()
plt.plot(ns,y)
plt.title("number of triangels as a function of n")
plt.show()
```

```python
# 12

ns = [int(i) for i in np.linspace(50, 1200, 24)]
y = []
for n in ns:
    p = 1 / n
    graphs = []
    for i in range(100):
        graphs.append(nx.fast_gnp_random_graph(n, p))
    triangles = []
    for g in graphs:
        triangles.append(sum(nx.triangles(g).values()) / 3)
    y.append(sum(triangles) / 100)

plt.figure()
plt.plot(ns,y,'-o')
plt.title("number of triangles as a function of n")
plt.show()
```

```python
#cumulative

#making cumulative sum
cumulative = np.cumsum(y)
i = 0
for n in ns:
    cumulative[i] = cumulative[i]/(i+1)  #mean
    i += 1
    
plt.plot(ns,cumulative)
plt.plot()
plt.title("mean cumulative function as a function of n")
plt.show()
```
